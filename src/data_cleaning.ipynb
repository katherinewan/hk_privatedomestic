```python
# import libraries
import numpy as np
import pandas as pd
import re
```

```python
# read raw file
db = pd.read_csv('../data/raw/avg_rent_month.csv')
```

```python
db.info()
db.head()
```

Remarks:
1. Delete all the Remarks columns.

```python
# find all the remarks columns
remark_cols = [col for col in db.columns if re.search(r'Remarks', col, re.IGNORECASE)]

# drop the remarks columns
db.drop(columns=remark_cols, inplace=True)
```

```python
db.head()
```

Remarks:
1. The information of Dec-25, Jan-26, Feb-26 are provisional.
2. Only analysis Nov-24 to Nov-25 data.

```python
# Change Month column to datetime format
db['Month'] = pd.to_datetime(db['Month'], format='%b-%y')

# Delete row excluding the period Nov-24 to Oct-25
db = db[(db['Month'] >= pd.to_datetime('Nov-24', format='%b-%y')) & (db['Month'] <= pd.to_datetime('Oct-25', format='%b-%y'))]

# Format Month column to MM-YY string format
db['Month'] = db['Month'].dt.strftime('%m-%y')
```

```python
db.head(12)
```

```python
# save changes to cleaned file
db.to_csv('../data/cleaned/avg_rent_month_cleaned.csv', index=False)
```
