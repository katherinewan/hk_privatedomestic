```python
# import libraries
import numpy as np
import pandas as pd
import re
```

```python
# read raw file
db = pd.read_csv('../data/raw/avg_rent_month.csv')
```

```python
db.info()
db.head()
```

Remarks:
1. Delete all the Remarks columns.

```python
# find all columns with "Remarks" in the name
remark_cols = [col for col in db.columns if re.search(r'Remarks', col, re.IGNORECASE)]

# replace "Remarks" with "< 20 transactions" in those column names
db.rename(columns={col: col.replace('Remarks', '< 20 transactions') for col in remark_cols}, inplace=True)
```

```python
db.head()
```

```python
# Change Null values to False, and Z to True
db.replace({np.nan: False, 'Z': True}, inplace=True)
```

```python
# Convert columns with "< 20 transactions" to boolean data type
transaction_cols = [col for col in db.columns if '< 20 transactions' in col]
db[transaction_cols] = db[transaction_cols].astype(bool)
```

Remarks:
1. The information of Dec-25, Jan-26, Feb-26 are provisional, aren't going to use it.
2. Only analysis Nov-15 to Nov-25 data.

```python
# Change Month column to datetime format
db['Month'] = pd.to_datetime(db['Month'], format='%b-%y')

# Delete row excluding the period Nov-24 to Oct-25
db = db[(db['Month'] >= pd.to_datetime('Nov-15', format='%b-%y')) & (db['Month'] <= pd.to_datetime('Nov-25', format='%b-%y'))]

# Format Month column to MM-YY string format
db['Month'] = db['Month'].dt.strftime('%m-%y')
```
